```python
#remove emoji
import emoji
def give_emoji_free_text(text):
    allchars = [str for str in text]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text = ' '.join([str for str in text.split() if not any(i in str for i in emoji_list)])
    return clean_text
```

```python
Translators=Translator()
data_list=[]
for i in data_df.index:
    if data_df['language'][i]:
        s=data_df['text'][i]
        s=give_emoji_free_text(s)
        s=Translators.translate(s).text
        data_list.append(s)
```

```python
data_df['label'] = data_df['Sentiment'].map({'neutral': 0, 'positive': 1,'negative':-1})
data_df.head()
```

```python
#Preprocessing text features
# Convert to lowercase
data_df['text_new']= data_df['text_new'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data_df.head()
```

```python

# Remove punctuations
data_df['text_new']= data_df['text_new'].str.replace('[^\w\s]','')
data_df['text_new'].head()
```

```python
# Remove stopwords
stop = stopwords.words('english')
data_df['text_new']= data_df['text_new'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data_df['text_new'].head()
```

```python
# Apply Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
data_df['text_new']= data_df['text_new'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
data_df['text_new'].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
train_counts= vectorizer.fit_transform(data_df['text_new'])
```

```python
#preprocess text by removing stopword and transform to numeric dataframe using regexp
token = RegexpTokenizer(r'[a-zA-Z]+')
cv = tfidfVectorizer(lowercase=True,stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
train_counts= cv.fit_transform(data_df['text_new'])
print(type(train_counts[0]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_counts, data_df['label'], test_size=0.25, random_state=1)
```

```python
#svm classifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
clf = SVC(gamma='scale', C=1000.0)
clf_svc=clf.fit(X_train,y_train)
predicted_svc=clf_svc.predict(X_test)
#Accuracy using svm Classifier
print("Accuracy:",accuracy_score(y_test, predicted_svc))
```

```python
import math
def normalize(score, alpha=15):
    """
    Normalize the score to be between -1 and 1 using an alpha that
    approximates the max expected value
    """
    norm_score = score/math.sqrt((score*score) + alpha)
    return norm_score
```

```python
compound=[]
for i in range(len(predicted_svc)):
    compound.append(normalize(predicted_svc[i]))
```

```python
# fit model and training data 
import xgboost
from xgboost import XGBClassifier
clf= XGBClassifier()
clf_xgb=clf.fit(X_train,y_train)
predicted_xgb=clf_xgb.predict(X_test)
print("Accuracy:",accuracy_score(y_test, predicted_xgb))
```
